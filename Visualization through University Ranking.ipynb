```python
# import packages

# Data Analysis Packages
import pandas as pd
import numpy as np

# Visualization Packages
from ggplot import *
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline

# Machine Learning Packages
from sklearn.model_selection import train_test_split
```

```python
uni = pd.read_csv('cwurData.csv')
```

```python
uni.info()
```

```python
# plot a correlation heatmap
plt.subplots(figsize = (15,15))
ax = plt.axes()
ax.set_title("University Ranking Correlation Heatmap")
corr = uni.corr()
corr
sns.heatmap(corr,
           xticklabels = corr.columns.values,
           yticklabels = corr.columns.values,
           cmap = "Reds")
```

```python
# Create a facet plot in ggplot
plot = ggplot(aesthetics = aes(x = 'quality_of_education', y = 'score', color = 'country', size = 'world_rank'), data = uni) + geom_point() 
plot += facet_grid('year')
plot += labs(title = 'University Ranking',
            x = 'Education Quality',
            y = 'Overall score')
print(plot)
```

```python
# test & train split
y = uni.score
X = uni.drop('score', axis = 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 123, stratify = y)
```
